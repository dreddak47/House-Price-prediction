```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#open xlsx file and read the data
df = pd.read_excel(f'GVSS_Vision_Data\Annotation.xlsx')
df = df.dropna()
print(df)
```

```python
#normalize the area and pincode columns
type(df['Area'])
#convert to numpy array 
area = df['Area'].to_numpy()
area=(area-np.min(area))/(np.max(area)-np.min(area))
print(area)


```

```python
pincode = df['Pincode']
#generte unique values for pincode
pincode = pincode.unique()
#sort the pincodes
pincode = np.sort(pincode)
mp={pincode:i for i,pincode in enumerate(pincode)}
mp
```

```python
pincode=[mp[i] for i in df['Pincode']]
df['Pincode']=pincode
df['Area']=area
```

```python
df.head()
```

```python
#write data to a new xlsx file
df.to_excel('normalized_data.xlsx',index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
corr = df.corr()
corr
#plot the correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()

```

```python
new_dataset_folder = 'GVSS_Vision_Data/Dataset_Final'
output_folder = 'GVSS_Vision_Data/Dataset_Final_grs'

import os
import shutil
import cv2

#make the output folder if not exists
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# for every image in the new_dataset_folder, grayscale it and histogram equalize it
for file in os.listdir(new_dataset_folder):
    if file.endswith('.jpg'):
        # read the image
        img = cv2.imread(os.path.join(new_dataset_folder, file))
        # convert the image to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        # histogram equalization
        equ = cv2.equalizeHist(gray)
        # save the image
        cv2.imwrite(os.path.join(output_folder, file), equ)
        print(f'Processed {file}')

    
```
